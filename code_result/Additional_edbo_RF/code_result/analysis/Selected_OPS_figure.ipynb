```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
model_list = [
    #"TrABsearch_top5_e0_ne5",
    #"BOsearch_EI",
    #"BOsearch_PI",
    "BOsearch_UCB", 
    "LRsearch",
    "RFsearch",
    #"RFsearch_e10%",
    #"RFsearch_e100%",
    #"TrABsearch_all_e0_ne5",
    #"TrABsearch_bottom5_e0_ne5",
    #"TrABsearch_top4_e0_ne5",
    "TrABsearch_top3_e0_ne5",
    #"TrABsearch_top5_e0_ne3",
    #"TrABsearch_top5_e0_ne5",
    #"RFsearch_e10%",
    #"TrABsearch_top5_e0_ne5_LGBM",
    #"TrABsearch_top5_e0_ne5_MF",
    #"TrABsearch_top5_e0_ne5_XGB",
    #"TrABsearch_top5_e0_ne10",
    #"TrABsearch_top5_e10%_ne5",
    ]
```

```python
All_data_names = [
    'Reaction_OCH2F',
    'Reaction_P', 'Reaction_Si'
    ]

for target in All_data_names:
    for model in model_list:
        if model == 'TrABsearch_all_e0_ne5':
            file_name = f'TrABsearch_{target}_top14_e0_ne5_result'                      
        elif model == 'TrABsearch_bottom5_e0_ne5':
            file_name = f'TrABsearch_{target}_bottom5_e0_ne5_result'
        elif model == 'TrABsearch_top4_e0_ne5':
            file_name = f'TrABsearch_{target}_top4_e0_ne5_result'
        elif model == 'TrABsearch_top3_e0_ne5':
            file_name = f'TrABsearch_{target}_top3_e0_ne5_result'
        elif model == 'TrABsearch_all_e0_ne5':
            file_name = f'TrABsearch_{target}_top14_e0_ne5_result'
        elif model == 'TrABsearch_top5_e0_ne3':
            file_name = f'TrABsearch_{target}_top5_e0_ne3_result'
        elif model == 'TrABsearch_top5_e0_ne5':
            file_name = f'TrABsearch_{target}_top5_e0_ne5_result'
        elif model == 'TrABsearch_top5_e0_ne5_LGBM':
            file_name = f'TrABsearch_{target}_top5_e0_ne5_result'            
        elif model == 'TrABsearch_top5_e0_ne5_MF':
            file_name = f'TrABsearch_{target}_top5_e0_ne5_result'    
        elif model == 'TrABsearch_top5_e0_ne5_XGB':
            file_name = f'TrABsearch_{target}_top5_e0_ne5_result'
        elif model == 'TrABsearch_top5_e0_ne10':
            file_name = f'TrABsearch_{target}_top5_e0_ne10_result'
        elif model == 'TrABsearch_top5_e10%_ne5':
            file_name = f'TrABsearch_{target}_top5_e0.1_ne5_result'
        elif model == 'BOsearch_EI':
            file_name = f'BOsearch_{target}_result'   
        elif model == 'BOsearch_PI':
            file_name = f'BOsearch_{target}_result'  
        elif model == 'BOsearch_UCB':
            file_name = f'BOsearch_{target}_result'  
        elif model == 'RFsearch_e10%':
            file_name = f'RFsearch_{target}_result'
        elif model == 'RFsearch_e100%':
            file_name = f'RFsearch_{target}_result'
        else:
            file_name = f'{model}_{target}_result'

        df = pd.read_csv(f'./results/summary_OPSsearch/stats/{model}/{file_name}_OPSsearch_stats.csv')

        average_row = df[df['Stat'] == 'average'].iloc[0]
        x_values = average_row.drop(['Stat']).values
        y_values = list(range(1, len(x_values) + 1))
        
        plt.plot(x_values, y_values, marker='o', markersize=10, linewidth=3, linestyle='-', label=model)
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    #plt.xlim(0,60)
    plt.title(f'{target}')
    plt.ylabel('Selected OPS count')
    plt.xlabel('Step')
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
    #plt.grid()

    plt.savefig(f'./figures/{target}.pdf')
    plt.show()
```
